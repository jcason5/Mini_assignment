### How does housing price change in relation to area?

```python
#importing matplotlib pandas and numpy
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
#importing dataset to use 
data=pd.read_csv("Housing.csv")
```

```python
#finding statistical summary of all numerical values
data.describe()
```

```python
#using info to find all categorical and numerical values
data.info()
```

```python
#making a sctter plot to find the relation between price and area
plt.scatter(data['price'], data['area'])
plt.xlabel("Price")
plt.ylabel("Area")
plt.title("Price in relation to Area")
plt.show()
```
